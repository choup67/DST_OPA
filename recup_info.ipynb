## récupération des prix pour chaque pair

```python
#récupération des informations price pour chaque pair
#fait le 09/07/2025 à 14h40
resp = requests.get('https://api.binance.us/api/v3/ticker/price').json()
data_price = pd.DataFrame(resp)
data_price
```

#récupération des pairs

```python
#récupération des pairs
pairs_list = []
for i in range(len(data_price)):
    pairs_list.append(data_price['symbol'][i])

pairs_list
```

## récupération du prix moyen

```python
#récupération du prix moyen
def recup_avg_price(liste) :
  avgprice_data = []
  for pair in liste :
    avgprice = requests.get('https://api.binance.us/api/v3/avgPrice?symbol={}'.format(pair)).json()
    avgprice_data.append(avgprice)

  df_avgprice = pd.DataFrame(avgprice_data)
  df_avgprice['symbol'] = liste
  df_avgprice = df_avgprice.rename(columns={'price':'avg_price'})
  df_avgprice = df_avgprice.astype({'avg_price':'float'})
  return df_avgprice
```

```python
#test de la fonction
#fait le 09/07/2025 à 14h40
df_avgprice = recup_avg_price(pairs_list)
df_avgprice
```

## récupération des trades récents

```python
#récupération dans un dataframe des trades recent par pair
def recup_trades(pair) :
  trades = requests.get('https://api.binance.us/api/v3/trades?symbol={}'.format(pair)).json()
  trades = pd.DataFrame(trades)
  trades['symbol'] = pair
  return trades
```

```python
trades_data = pd.DataFrame({})

for pair in pairs_list:
    trades = recup_trades(pair)
    #display(trades.head(3))
    trades_data = pd.concat([trades_data, trades], ignore_index=True)

trades_data = pd.DataFrame(trades_data)
```

```python
from time import sleep
#définition de la fonction pour la récupération des infos kline
def recup_klines(pair):
    kline = requests.get('https://api.binance.us/api/v3/klines?symbol={}&interval=1m'.format(pair)).json()
    #définition des colonnes
    columns = [
        "open_time", "open", "high", "low", "close", "volume",
        "close_time", "quote_asset_volume", "number_of_trades",
        "taker_buy_base_asset_volume", "taker_buy_quote_asset_volume", "ignore"
    ]
    data = pd.DataFrame(kline, columns=columns)
    data["symbol"] = pair
    return data

# Récupération des données pour chaque symbole
all_data = []
for pair in pairs_list:
    df = recup_klines(pair)
    if not df.empty:
        all_data.append(df)
    sleep(0.5)

# Concaténation finale
if all_data:
    full_df = pd.concat(all_data, ignore_index=True)
    print(full_df.head())
else:
    print("Aucune donnée récupérée.")
```
